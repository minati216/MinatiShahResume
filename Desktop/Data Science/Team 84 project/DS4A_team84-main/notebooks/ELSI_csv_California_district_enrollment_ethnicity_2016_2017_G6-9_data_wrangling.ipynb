```python
df = pd.read_csv('commondata/ELSI_csv_California_district_enrollment_ethnicity_2016_2017_G6-9.csv', error_bad_lines=False, delimiter=',')
```

```python
df.iloc[:,2:] = df.iloc[:,2:].apply(pd.to_numeric, errors='coerce')
```

```python
df_grades = df.copy()
df_school_districts_by_race = df.copy().iloc[:,:1]
df_school_districts_by_race
```

```python
american_indian = ['Grade 6 Students - American Indian/Alaska Native [District] 2016-17', 
           'Grade 7 Students - American Indian/Alaska Native [District] 2016-17', 
           'Grade 8 Students - American Indian/Alaska Native [District] 2016-17',
           'Grade 9 Students - American Indian/Alaska Native [District] 2016-17']
asian = ['Grade 6 Students - Asian or Asian/Pacific Islander [District] 2016-17', 
           'Grade 7 Students - Asian or Asian/Pacific Islander [District] 2016-17', 
           'Grade 8 Students - Asian or Asian/Pacific Islander [District] 2016-17',
           'Grade 9 Students - Asian or Asian/Pacific Islander [District] 2016-17']

df_school_districts_by_race['American Indian/Alaska Native'] = df_grades[american_indian].sum(axis=1)
df_school_districts_by_race['Asian or Asian/Pacific Islander'] = df_grades[asian].sum(axis=1)
```
